```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris 

iris1 = load_iris()

print(iris1.target)
```

```python
# To create a dataframe

iris2 = pd.DataFrame(iris1.data, columns=iris1.feature_names)

print(iris2.describe())
```

```python
iris2['Target'] = iris1.target

print(iris2)
```

```python
import seaborn as sns

sns.displot(iris2.Target, kde=True)
plt.show()
```

```python
sns.pairplot(iris2, hue='Target')
plt.show()
```

```python
corr_matrix = iris2.corr()
sns.heatmap(corr_matrix, annot=True, cmap='viridis')
plt.show()
```

```python
#Feature scaling

from sklearn.preprocessing import MinMaxScaler

mm = MinMaxScaler()

mm1 = mm.fit_transform(iris1.data)
print(mm1)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(mm1, iris1.target, train_size=0.80)

print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
```

```python
from sklearn.naive_bayes import GaussianNB

gb = GaussianNB()

model = gb.fit(X_train, y_train)

print(model.score(X_train, y_train))
print(model.score(X_test, y_test))
```
